## Text Cleaning

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk
nltk.download('stopwords')
nltk.download('punkt')
```

```python
example_sent = "This is a sample> sentence, showing < off the stop words filtration."

punctuation = set(['.','<','>', '?', '!', ',', '$', ':', ';', '(',')','-',"`",'\'','"','>>','|','."',',"','@','#','%','^','*','&'])
stop_words = set(stopwords.words('english'))| punctuation

def clean_sentence(example_sent):
    word_tokens = word_tokenize(example_sent)
    filtered_sentence = [w for w in word_tokens if not w.lower() in stop_words]
    a = ( ' '.join(word for word in filtered_sentence))
    return a

print(clean_sentence(example_sent))
```

```python
from textblob import TextBlob as tb
```

```python
pros_list = list(df_req['Pros'])
tb(pros_list[1])
```

```python
pros_list = list(df_req['Pros'])
pros_cleaned_list = list(map(clean_sentence ,pros_list))
pros_cleaned_list[1]
```

## Forming TF-IDF Matrix

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer()

# CSR Matrix below will have the tf-idf vectors for all the Pros 
csr_mat = vectorizer.fit_transform(pros_cleaned_list)
print(vectorizer.get_feature_names())
print(csr_mat.shape)
print(type(csr_mat))
```

## Dimensional Reduction followed by Clustering

```python
# Dimensional reduction removes noise. Try with different count of components
# Clustering helps to identify topics. Try with different count of clusters
from sklearn.decomposition import TruncatedSVD
from sklearn.cluster import KMeans
from sklearn.pipeline import make_pipeline

# Create a TruncatedSVD instance: svd
svd = TruncatedSVD(n_components = 100)

# Create a KMeans instance: kmeans
kmeans = KMeans(n_clusters=8)

# Create a pipeline: pipeline
pipeline = make_pipeline(svd,kmeans)

# Fit the pipeline to articles
pipeline.fit(csr_mat)

# Calculate the cluster labels: labels
labels = pipeline.predict(csr_mat)

# Create a DataFrame aligning labels and titles: df
df_clusters = pd.DataFrame({'cluster': labels, 'article_clean': pros_cleaned_list,'article_original':pros_list})

# Display df sorted by cluster label
print(df_clusters.sort_values('cluster'))
```

```python
# Save the clusters
df_clusters.to_csv('PythonExport.csv', sep=',')
```

Each pro doesn't just talk about a single review. Instead it concentrates on multiple topics. For example, an employee might higly speak of both work-culture and salaries. But using KMeans or any other type of clustering, each review is forced under only one cluster. Hence, clustering isn't the ideal option here. Instead, we will follow a simpler approach where a review can fall under different topics. Identify the important words (based on all reviews) to form certain topics. If a review has at least one word from the important words that define a topic, this review falls under the corresponding topic.

## Identify the most important words

```python
bloblist = []
for sentence in pros_cleaned_list:
    blob = tb(sentence)
    bloblist.append(blob)
print(len(bloblist))
print(bloblist[0])
```

```python
def tf(word, blob):
    return blob.words.count(word) / len(blob.words)

def n_containing(word, bloblist):
    return sum(1 for blob in bloblist if word in blob)

def idf(word, bloblist):
    return math.log(len(bloblist) / (1 + n_containing(word, bloblist)))

def tfidf(word, blob, bloblist):
    return tf(word, blob) * idf(word, bloblist)
```

```python
for i, blob in enumerate(bloblist):
    print("Top words in document {}".format(i + 1))
    scores = {word: tfidf(word, blob, bloblist) for word in blob.words}
    sorted_words = sorted(scores.items(), key=lambda x: x[1], reverse=True)
    for word, score in sorted_words[:3]:
        print("Word: {}, TF-IDF: {}".format(word, round(score, 5)))
```

```python
all_words =[]
for i, blob in enumerate(bloblist):
    print("Top words in document {}".format(i + 1))
    scores = {word: tfidf(word, blob, bloblist) for word in blob.words}
    sorted_words = sorted(scores.items(), key=lambda x: x[1], reverse=True)
    all_words.append(sorted_words)
    print("sorted_words "+str(len(sorted_words)))
    print("all_words "+str(len(all_words)))
```

```python
# all_words will have sorted words of all documents (each doc as an element)
print(len(all_words))
print(all_words[-1])
```
